```python
import pandas as pd
import numpy as np

from collections import Counter
```

```python
MIN_INFLUENCED = 0
MAX_INFLUENCED = 18470
Q = 5
```

```python
df = pd.read_csv('parallel_merge_plurality.csv', names=['k','node','number_influenced','s'])
```

```python
df.groupby(['node','s'], sort=False)['k'].max()
idx = df.groupby(['node','s'])['k'].transform(max) == df['k']
df_plurality_max_k = df[idx][['node','number_influenced','s']]
```

```python
df_plurality_max_k = df_plurality_max_k.reset_index(drop=True)
df_plurality_max_k.head()
```

```python
categories = pd.qcut(range(MIN_INFLUENCED, MAX_INFLUENCED + 2), Q)
category_converted = [categories.categories.right[i] for i in range(Q)]
category_converted
```

```python
results_merge = []
```

```python
for s in df_plurality_max_k['s'].unique():
    
    results = {}
    results['name'] = 'plurality'
    results['level l'] = s
    df_slide = df_plurality_max_k.loc[df_plurality_max_k['s'] == s].copy()
    
    results['min_value'] = df_slide['number_influenced'].min()
    results['max_value'] = df_slide['number_influenced'].max()
    results['mean'] = df_slide['number_influenced'].mean()
    results['std'] = df_slide['number_influenced'].std()
    
    categorized_numbers = np.digitize(df_slide['number_influenced'], category_converted)
    dict_counter = Counter(categorized_numbers)
    
    for x in dict_counter.keys():
        results['q{}'.format(x+1)] = dict_counter[x]
        
    results_merge.append(results)
```

```python
pd.DataFrame(results_merge).fillna(0).round(2)[['name','level l','max_value','min_value','mean','std','q1','q2','q3','q4','q5']]
```
